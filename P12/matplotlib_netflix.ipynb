```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df = df[df['type'] == 'Movie']
df['duration'] = df['duration'].str.extract('(\d+)').astype(float)
avg = df.groupby('country')['duration'].mean().head(10)

avg.plot(kind='bar')
plt.xticks(rotation=45)
plt.ylabel('Average Duration (minutes)')
plt.title('Average Movie Duration by Country (Top 10)')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
movies_df = df[df['type']== 'Movie']
movies_df['duration'] = df['duration'].str.extract('(\d+)').astype(float)
avg_duration_by_country = movies_df.groupby('country')['duration'].mean().sort_values(ascending=False)
avg_duration_by_country.head(10).plot(kind='bar',figsize=(10,6),color='blue')

plt.title('Top 10 Countries by average MOvie Duration')
plt.xlabel('country')
plt.ylabel('average Duration(minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
